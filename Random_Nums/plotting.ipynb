```python
import numpy as np, matplotlib.pyplot as plt
%matplotlib inline
x, d = np.loadtxt("density.csv", delimiter=",", unpack=True)
plt.bar(x, d, width=0.05, alpha=0.5, label="sampled")
plt.plot(x, 2*np.exp(-2*x), 'r-', label="theory: $\\lambda e^{-\\lambda x},\\ \\lambda=2$")
plt.xlabel("Free path length x")
plt.ylabel("Probability density p(x)")
plt.legend()
plt.title("Monte Carlo free-path sampling vs. analytic PDF")
plt.show()
```
